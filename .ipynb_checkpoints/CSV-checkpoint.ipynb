```python
# USING CSV package instead of pandas
import csv
fields=['Kamal','23','Baluwatar']
with open('books.csv','a') as f:
    x = csv.writer(f)
    x.writerow(fields)
```

```python
import csv
fields=[['Kamal','23','Baluwatar'],['Ranjan','22','Tokha']]
with open('books.csv','a') as f:
    x = csv.writer(f)
    for i in fields:
        x.writerow(i)
```
